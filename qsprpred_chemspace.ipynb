```python
gamma=1
test_size=0.2
sim_threshold=0.25
```

```python
from qsprpred.data.sources.papyrus import Papyrus

acc_keys = ["P41597"]
dataset_name = "CCR2_HUMAN"  # name of the file to be generated
quality = "high"  # choose minimum quality from {"high", "medium", "low"} for the standard set
papyrus_version = '05.6'  # Papyrus database version
data_dir = "data"

papyrus = Papyrus(
    data_dir=data_dir,
    version=papyrus_version,
    stereo=False,
    plus_only=False, #  set this to True  if you only  want cleaned high quality data
    descriptors=None
)

mt = papyrus.getData(
    acc_keys,
    quality,
    name=dataset_name,
    use_existing=True
)
mt.getDF()
```

```python
from qsprpred.models.tasks import TargetTasks
from qsprpred.data.data import QSPRDataset

target_props=[{"name": "pchembl_value_Median", "task": TargetTasks.SINGLECLASS, 'th': [7]}]
ds = QSPRDataset.fromMolTable(mt, target_props=target_props)
ds.targetProperties
```

```python
from qsprpred.data.utils.descriptorcalculator import MoleculeDescriptorsCalculator
from qsprpred.data.utils.descriptorsets import FingerprintSet
desc_calculator = MoleculeDescriptorsCalculator(desc_sets=[FingerprintSet(fingerprint_type="MorganFP", radius=3, nBits=2048)])
ds.addDescriptors(desc_calculator, recalculate=False)
```

```python
from tools import AnalogueSplit

split = AnalogueSplit(
    gamma=gamma, 
    test_size=test_size, 
    descriptor="morgan3", 
    sim_threshold=sim_threshold, 
    random_seed=42
)
ds.split(split, featurize=True)
```

```python
ds.getDF()["IsTrain"] = ds.getDF().index.isin(ds.X.index)
ds.getProperty("IsTrain")
```

```python
from scaffviz.clustering.manifold import TSNE
from scaffviz.depiction.plot import Plot


plt = Plot(TSNE())
_ = plt.plot(
    ds,
    color_by='IsTrain',
    recalculate=False,
    card_data=[
        'pchembl_value_Median',
        'IsTrain'
    ],
    interactive=True,
    port=9112
)
```
