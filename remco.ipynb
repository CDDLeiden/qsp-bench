```python
import pandas as pd

df_all = pd.read_table('data/papyrus-05.5_targets_bioactivity.txt', header=0)
df_all
```

```python
from qsprpred.models.tasks import TargetTasks
from qsprpred.data.data import QSPRDataset

target_props = [{"name": "pchembl_value_Median", "task": TargetTasks.SINGLECLASS, 'th': [6.5]}]
data_settings = []
for target, data in df_all.groupby('accession'):
    name = f"{target}_remco"
    ds = QSPRDataset(name=name, df=data, store_dir="./data", target_props=target_props, overwrite=True)
    ds.save()
    data_settings.append(
        {
            "acc_keys": [target],
            "dataset_name": name,
            "quality": "high",
            "papyrus_version": '05.6',
            "plus_only": True,
            "activity_types": "all"
        }
    )
data_settings
```

```python
import json

json.dump(data_settings, open("./data/remco.json", "w"))
```
